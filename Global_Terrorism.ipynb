# **DATA CLEANING**

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','latitude':'Latitude','longitude':'Longitude','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'Attacktype','targtype1_txt':'Target_type','targsubtype1_txt':'Targetsubtype','weaptype1_txt':'Weapontype','weapsubtype1_txt':'Weaponsubtype','motive':'Motive','gname':'Group','nkill':'Killed','nwound':'Wounded','propextent_txt':'property damage'},inplace=True)
```

```python
df1=df[['Year','Month','Day','Latitude','Longitude','Country','Region','State','city','Attacktype','Target_type','Targetsubtype','Weapontype','Weaponsubtype','Motive','Group','Wounded','Killed','property damage']]
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1.shape
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
df1['Killed']=df1['Killed'].fillna(0)
df1['Wounded']=df1['Wounded'].fillna(0)
df1['Casualties']=df1['Killed']+df1['Wounded']
```

```python
df1.describe()
```

# **visualization**


# Number of attacks in each year



```python
attacks_count=df1['Year'].value_counts().sort_index()
attacks_count
```

```python
plt.figure(figsize=(25,8))
attacks_count.plot.bar(color='Red')
plt.title('Timeline of Attacks',fontsize=24)
plt.xlabel('Year',fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.ylabel('No of attacks',fontsize=24)
plt.show()
```

# country that suffered the most attacks 



```python
df1['Country'].value_counts()
```

```python
plt.subplots(figsize=(25,8))
sns.countplot(x='Country',data=df1,order=df1['Country'].value_counts().index[:50])
plt.title('Number of attacks in country',fontsize=24)
plt.xlabel('country',fontsize=24)
plt.ylabel('No of attacks',fontsize=24)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15,)
plt.show()
```

# state that suffered most attacks

```python
df1['State'].value_counts().drop('unknown')
```

```python
plt.subplots(figsize=(25,8))
sns.countplot(x='State',data=df1,palette='RdPu',order=df1['State'].value_counts().index[:50])
plt.title('number of attacks in state ',fontsize=24)
plt.xlabel('State',fontsize=24)
plt.ylabel('No of attacks',fontsize=24)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15,)
plt.show()
```

# Number of attacks in each city 

```python
df1['city'].value_counts()
```

```python
plt.subplots(figsize=(25,8))
sns.countplot(x='city',data=df1,palette='RdYlBu',order=df1['city'].value_counts().index[1:30])
plt.title('Number of attacks in  city',fontsize=24)
plt.xlabel('city',fontsize=24)
plt.ylabel('No of attacks',fontsize=24)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

# Region Wise Attacks over the period 1970-2017




```python
region_attack=pd.crosstab(df1.Year,df1.Region)
region_attack
```

```python
region_attack.plot(kind='area',figsize=(25,12))
plt.title('Terrorist Attacks by region',fontsize=24)
plt.xlabel('Year',fontsize=24)
plt.ylabel('NOof Attacks',fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

# Type of attacks

```python
attack_counts = df1['Attacktype'].value_counts()
attack_counts
```

```python
plt.figure(figsize=(6,6))
plt.pie(attack_counts,labels=attack_counts.index,autopct='%23.1f%%',textprops={'fontsize':9}, startangle=180)
plt.title('Attack Type Distribution')
plt.axis('equal')
plt.legend(attack_counts.index,loc='upper right',bbox_to_anchor = (1.4,0.5,0.4,0.4))
plt.show()
```

# Types of Weapons used in Attacks

```python
plt.subplots(figsize=(25,8))
sns.barplot(x=df['Attacktype'].value_counts()[:10].index,y=df['Attacktype'].value_counts()[:10].values,palette = 'mako')
plt.xlabel('Weapon',fontsize=24)
plt.ylabel('No of attack',fontsize=24)
plt.xticks(fontsize=14)
plt.yticks(fontsize=18)
plt.title('Weapons used to attacks',fontsize=18)
plt.xticks(rotation = 50)
plt.show()
```

# Group wise Attack

```python
#ga denotes the group wise attacks
ga=df1['Group'].value_counts()[1:10]
ga
```

# Casualties in the most frequent Attack types

```python
#ac represents the casualties in an attack type
cas=df1[["Attacktype","Casualties"]].groupby("Attacktype").sum().sort_values(by="Casualties",ascending=False)
```

```python
cas.plot(kind="bar",color="Purple",figsize=(15,6))
plt.title("Casualties in each Attack",fontsize=13)
plt.xlabel("Attacktypes",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

# Casualties in various Target types

```python
cas1=df1[["Target_type","Casualties"]].groupby("Target_type").sum().sort_values(by="Casualties",ascending=False)
```

```python
cas1.plot(kind="bar",color="Blue",figsize=(15,6))
plt.title("Casualties in each Target Attack",fontsize=13)
plt.xlabel("Target Types",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```
